# 11 - Refazendo Dummies

O processamento dos Dummies anteriormente apresentou resultados inferiores no teste (72%). Será feito um pré processamento dos dados tratando os andares de forma diferente.

Criando o dataset com o máximo de informações possivelmente válida.

## Preparando o ambiente

```python
import pandas as pd
```

## Carregando os dados

### Treino

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/SalatielBairros/kaggle-titanic/main/data/processed/train_complete.csv')
titanic.sample(2)
```

```python
titanic.drop(columns=['Unnamed: 0', 'PassengerId', 'Name', 'Ticket', 'title', 'faixa', 'Floor'], inplace=True)
titanic.sample(2)
```

```python
map_cabine = {
    False: 0,
    True: 1
}
titanic['possui_cabine'] = titanic['possui_cabine'].map(map_cabine)
```

```python
titanic.sample(2)
```

```python
dummies = pd.get_dummies(titanic)
dummies.sample(5)
```

### Teste

```python
p_teste = pd.read_csv('https://raw.githubusercontent.com/SalatielBairros/kaggle-titanic/main/data/processed/test_processed.csv')
o_teste = pd.read_csv('https://raw.githubusercontent.com/SalatielBairros/kaggle-titanic/main/data/original/test.csv')
p_teste['Embarked'] = o_teste['Embarked']
p_teste['Fare'] = o_teste['Fare']
p_teste.drop(columns=['Unnamed: 0', 'Floor', 'acompanhado', 'faixa_etaria'], inplace=True)
p_teste['possui_cabine'] = p_teste['possui_cabine'].map(map_cabine)
t_dummies = pd.get_dummies(p_teste)
t_dummies.sample(5)
```

```python
print(dummies.shape)
print(t_dummies.shape)
```

## Salvando os dados

```python
dummies.to_csv('../../data/processed_v2/train_dummies.csv')
t_dummies.to_csv('../../data/processed_v2/test_dummies.csv')
```
